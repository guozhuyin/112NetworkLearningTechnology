# 檔案、資料檔或資料庫的讀寫

```python
# 讀取數據集
data=pd.read_csv("studentPerformance.csv")
data
```

# 資料的排序、篩選或增修刪查

```python
#利用 pandas中 isnull().sum()查詢缺失的過濾
data.isnull().any().sum()
```

```python
data.describe()
```

```python
# 計算總分
total_score = (data["math score"] + data["reading score"] + data["writing score"])
average_total_score = ((data["math score"] + data["reading score"] + data["writing score"])/3).round()
# 新增總分欄位
data["average_total_score"] = average_total_score
# 排序總分由大至小
data = data.sort_values("average_total_score", ascending=False)
data
```

# 迴圈及判斷指令

```python
# 將60分及格以上的人與不及格的人做區分
for i in range(len(data)):
    if data["average_total_score"][i] >= 60:
        data.loc[i, "group"] = "pass"
    else:
        data.loc[i, "group"] = "fail"
pass_group = data["group"] == "pass"
fail_group = data["group"] == "fail"

print("Pass的總數:", pass_group.sum())
print("Fail的總數:", fail_group.sum())        
data

labels = ["Pass", "Fail"]
sizes = [pass_group.sum(), fail_group.sum()]

plt.pie(sizes, labels=labels, autopct="%1.0f%%")
plt.show()
```

# 統計分析 & 視覺化分析

```python
# 計算總體平均分、標準差、分位數等統計量
mean_score = data["average_total_score"].mean().round()

q25_score = data["average_total_score"].quantile(0.25).round()
q75_score = data["average_total_score"].quantile(0.75).round()
print("\n","mean_score = ",mean_score,"\n","q25_score = ",q25_score,"\n","q75_score = ",q75_score)

plt.hist(data["average_total_score"], bins=50, range=(0, 100))
plt.axvline(mean_score, color="red", linestyle="-")
plt.axvline(q25_score, color="red", linestyle="--")
plt.axvline(q75_score, color="red", linestyle="--")
plt.xlabel("score")
plt.ylabel("student number")
plt.show()
```

# 機器學習
- 均方誤差 (Mean Squared Error, MSE)：MSE是迴歸問題中常用的指標，它衡量模型預測值與真實值之間的平均平方差。 MSE越小，表示模型的預測越準確。
- R平方 (R-squared, R2)：R2衡量了模型對目標變數變異數的解釋程度。 R2的取值範圍在0到1之間，越接近1表示模型越好。


# DecisionTreeClassifier
